# Problema 1 - Confeitaria

```python
prob = LpProblem("Ex1", LpMaximize)
x1 = LpVariable("x1", 0)
x2 = LpVariable("x2", 0)

# Função Objetivo
prob += 20 * x1 + 15 * x2

# Restrições
prob += 2 * x1 + x2 <= 20   # Farinha (kg)
prob += 3 * x1 + 3 * x2 <= 36 # Ovos (unid.)
prob += x1 + 2 * x2 <= 16   # Tempo (h)

prob.solve()

print("Status da Solução:", LpStatus[prob.status])
print("-" * 30)
print("Plano de Produção Ideal:")
print(f"Quantidade de Chocolate (x1): {x1.varValue}")
print(f"Quantidade de Baunilha  (x2): {x2.varValue}")
print("-" * 30)
print(f"Lucro Máximo (Z): R$ {pulp.value(prob.objective)}")
```

# Problema 6 - Fábrica de Tintas Resindencias

```python
prob = LpProblem("Prob6_Tintas_Residenciais", LpMaximize)
x1 = LpVariable("x1", 0)
x2 = LpVariable("x2", 0)
 
# Função Objetivo
prob += 10*x1 + 8*x2
 
# Restrições
prob += 6 * x1 + 4 * x2 <= 24  # Pigmento M1 (kg)
prob += x1 + 2 * x2 <= 6     # Solvente M2 (L)
prob += x2 - x1 <= 1       # Demanda relativa (interiores fachadas)
prob += x2 <= 2            # Demanda máxima de interiores (galões)
prob += x1 + x2 <= 5       # Produção total (galões)
 
prob.solve()
 
print("\nStatus da Solução:", LpStatus[prob.status])
print("-" * 30)
print("Plano de Produção Ideal:")
print(f"Quantidade de Tinta Fachadas   (x1): {x1.varValue}")
print(f"Quantidade de Tinta Interiores (x2): {x2.varValue}")
print("-" * 30)
print(f"Lucro Máximo (Z): R$ {pulp.value(prob.objective)}")
```

# Problema 8 - Produção de Dispositivos Eletrônicos

```python
prob = LpProblem("Prob8_Dispositivos_Eletronicos", LpMaximize)
x1 = LpVariable("x1", 0)
x2 = LpVariable("x2", 0)
 
# Função Objetivo
prob += 100*x1 + 120*x2
 
# Restrições
prob += 4 * x1 + 2 * x2 <= 40  # Montagem (h)
prob += x1 + 2 * x2 <= 16    # Teste de qualidade (h)
prob += x2 <= 2 * x1         # Diversidade (smartphones <= 2x tablets)
prob += x1 <= 8             # Máximo de tablets (unid.)
prob += x1 + x2 >= 5       # Total mínimo (unid.)
 
prob.solve()
 
print("\nStatus da Solução:", LpStatus[prob.status])
print("-" * 30)
print("Plano de Produção Ideal:")
print(f"Quantidade de Tablets     (x1): {x1.varValue}")
print(f"Quantidade de Smartphones (x2): {x2.varValue}")
print("-" * 30)
print(f"Lucro Máximo (Z): R$ {pulp.value(prob.objective)}")
```

# Problema 9 - Mix de Produção de Móveis Modulados

```python
prob = LpProblem("Prob9_Moveis_Modulados", LpMaximize)
x1 = LpVariable("x1", 0)
x2 = LpVariable("x2", 0)
x3 = LpVariable("x3", 0)
 
# Função Objetivo
prob += 150*x1 + 100*x2 + 80*x3
 
# Restrições
prob += 10 * x1 + 5 * x2 + 4 * x3 <= 200  # MDF (m^2)
prob += 6 * x1 + 3 * x2 + 2 * x3 <= 120   # Montagem (h)
prob += x1 >= 5                       # Mínimo de armários
prob += x2 >= 5                       # Mínimo de estantes
prob += x3 >= 5                       # Mínimo de gaveteiros
prob += x1 <= 15                      # Máximo de armários
prob += x2 + x3 <= 30                 # Logística de entrega
 
prob.solve()
 
print("\nStatus da Solução:", LpStatus[prob.status])
print("-" * 30)
print("Plano de Produção Ideal:")
print(f"Quantidade de Armários   (x1): {x1.varValue}")
print(f"Quantidade de Estantes   (x2): {x2.varValue}")
print(f"Quantidade de Gaveteiros (x3): {x3.varValue}")
print("-" * 30)
print(f"Lucro Máximo (Z): R$ {pulp.value(prob.objective)}")
```

# Problema 10 - Otimização de DIeta para Gado de Corte

```python
prob = LpProblem("Prob10_Dieta_Gado_de_Corte", LpMinimize)
x1 = LpVariable("x1", 0)
x2 = LpVariable("x2", 0)
x3 = LpVariable("x3", 0)
 
# Função Objetivo
prob += 2*x1 + 5*x2 + 12*x3
 
# Restrições
prob += 5 * x1 + 20 * x2 + 10 * x3 >= 400  # Proteína (unid.)
prob += 30 * x1 + 10 * x2 + 5 * x3 >= 600  # Energia (unid. calóricas)
prob += x1 + x2 + x3 <= 100          # Peso total da mistura (kg)
prob += x1 + x2 - 9 * x3 >= 0          # Suplemento <= 10% da mistura
prob += x1 - 2 * x2 >= 0               # Palatabilidade (milho >= 2x soja)
 
prob.solve()
 
print("\nStatus da Solução:", LpStatus[prob.status])
print("-" * 30)
print("Plano de Dieta Ideal:")
print(f"Quantidade de Milho           (x1): {x1.varValue}")
print(f"Quantidade de Farelo de Soja  (x2): {x2.varValue}")
print(f"Quantidade de Suplemento Min. (x3): {x3.varValue}")
print("-" * 30)
print(f"Custo Mínimo (Z): R$ {pulp.value(prob.objective)}")
```
